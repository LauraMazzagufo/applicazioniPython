```python
import os
import json
from google.colab import files
from PIL import Image

# CONFIG GENERICA
base_url = "https://img.serverchroma.cnr.it/iiif"
manifest_base_url = f"{base_url}/iiif-manifests"
image_base_path = "pavese%2Fromanzi%2F{filename}"  # path URL-encoded

# CHIEDI IL NOME DEL CAPITOLO DE LA LUNA E I FALO
capitolo_name = input("Inserisci il nome del file indicando il capitolo e il numero in formato arabo, ad esempio 'capitolo 1': ").strip()
capitolo_id = capitolo_name.lower().replace(" ", "")
manifest_id = f"LF_{capitolo_id}_manifest"
manifest_url = f"{manifest_base_url}/{manifest_id}.json"

# UPLOAD FILE
uploaded = files.upload()

# ORDINA FILES
filenames = sorted(uploaded.keys())

# CREA MANIFEST
manifest = {
    "@context": "http://iiif.io/api/presentation/2/context.json",
    "@id": manifest_url,
    "@type": "sc:Manifest",
    "label": f"La luna e i falò: '{capitolo_name}'",
    "description": f"La luna e i falò, '{capitolo_name}', edizione critica a cura di Miryam Grasso, Edizioni Sinestesie, Avellino, 2020.",
    "license": "https://creativecommons.org/publicdomain/mark/1.0/",
    "attribution": "ISTC - CNR",
    "sequences": [
        {
            "@type": "sc:Sequence",
            "label": "Sequence 1",
            "canvases": []
        }
    ]
}

# CREA CANVAS PER OGNI IMMAGINE
for idx, filename in enumerate(filenames, start=1):
    img = Image.open(filename)
    width, height = img.size
    base_file = os.path.basename(filename)
    file_no_ext = base_file.replace(".png", "")

    # Rimuove "LF_" se presente
    file_id = file_no_ext
    if file_id.startswith("LF_"):
        file_id = file_id[3:]

    # Pulisce il nome per label e descrizione
    file_label_clean = file_id.replace("_", " ")
    label = f"Pagina {idx}: {file_label_clean}"
    description_text = f"Source: {file_label_clean}"

    encoded_path = image_base_path.format(filename=base_file)
    canvas_id = f"{base_url}/iiif/canvas/{manifest_id}_{idx}"

    # Composizione URL related resource
    poetry_url_title = capitolo_name.replace(" ", "_")
    related_resource_url = (
        f"https://digitalpavese.cnr.it/pavese-in-testo/narrativa"
    )

    canvas = {
        "@id": canvas_id,
        "@type": "sc:Canvas",
        "label": label,
        "height": height,
        "width": width,
        "description": description_text,
        "metadata": [
            {
                "label": "current element",
                "value": description_text
            },
            {
                "label": "related resources",
                "value": f'<a href="{related_resource_url}">Edizione in PAVES-e</a>'
            }
        ],
        "thumbnail": {
            "@id": f"{base_url}/3/{encoded_path}/full/200,/0/default.jpg",
            "format": "image/jpeg"
        },
        "images": [
            {
                "@id": f"{canvas_id}/annotation",
                "@type": "oa:Annotation",
                "motivation": "sc:painting",
                "on": canvas_id,
                "resource": {
                    "@id": f"{base_url}/3/{encoded_path}/full/max/0/default.jpg",
                    "@type": "dctypes:Image",
                    "format": "image/jpeg",
                    "height": height,
                    "width": width,
                    "service": {
                        "@id": f"{base_url}/3/{encoded_path}",
                        "profile": "http://iiif.io/api/image/2/level2.json",
                        "@context": "http://iiif.io/api/image/2/context.json"
                    }
                }
            }
        ]
    }

    manifest["sequences"][0]["canvases"].append(canvas)

# SALVA E SCARICA MANIFEST
output_filename = f"{manifest_id}.json"
with open(output_filename, "w") as f:
    json.dump(manifest, f, indent=2)

files.download(output_filename)
```
